```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://stardewvalleywiki.com/Fish#Fishing_Pole_Fish"  # Replace with the actual URL
headers = {"User-Agent": "Mozilla/5.0"}  # Helps avoid blocks
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, "lxml")  # Parse HTML
else:
    print(f"Failed to fetch page. Status code: {response.status_code}")
```

```python
table = soup.find("table")
```

```python
# Extract headers
headers = [th.text.strip() for th in table.find_all("th")]
print(f"Extracted {len(headers)} headers: {headers}")

# Extract rows
data = []
rows = table.find_all("tr")
print(f"Found {len(rows)} rows")  # Debugging row count

for i, row in enumerate(rows[1:], start=1):  # Skip the header row
    cells = row.find_all(["td", "th"])  # Capture both <td> and <th> in rows
    cell_text = [cell.text.strip() for cell in cells]
    
    # Debugging: Print each row's content and column count
    print(f"Row {i}: {cell_text} (Columns: {len(cell_text)})")
    
    # Only append rows with data
    if len(cell_text) > 0:
        data.append(cell_text)

# Ensure all rows have the same number of columns as headers
max_cols = max(len(row) for row in data)
if len(headers) != max_cols:
    print(f"WARNING: Header count ({len(headers)}) does not match max row count ({max_cols})")
    headers = [f"Column {i+1}" for i in range(max_cols)]  # Generate generic column names

df = pd.DataFrame(data, columns=headers)
df.head(20)
```

```python
useful_rows_df = df.iloc[::13]

useful_rows_df.head(3)
```

```python
useful_rows_df = useful_rows_df.drop(useful_rows_df.columns[list(range(2, 30)) + [0] + [37] + [36] + [35] + [34]], axis=1)
useful_rows_df = useful_rows_df.reset_index(drop=True)
useful_rows_df.head(3)
```

```python
useful_rows_df.to_excel("filtered_table_data.xlsx", index=False, header=False)
```
