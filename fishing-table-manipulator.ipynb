```python
import pandas as pd

df = pd.read_excel("filtered_table_data.xlsx")
df.head()
```

```python
# Define valid season names
valid_seasons = {"Spring", "Fall", "Summer", "Winter", "All Seasons"}

# Function to correctly split the Season column
def split_seasons(season):
    # Ensure "All Seasons" remains intact
    if season == "All Seasons":
        return [season]  # Return as a single-element list
    return season.split()  # Split all others

# Apply the function and explode
df["Season"] = df["Season"].apply(split_seasons)
df_exploded = df.explode("Season").reset_index(drop=True)

# Filter to keep only valid seasons
df_filtered = df_exploded[df_exploded["Season"].isin(valid_seasons)].reset_index(drop=True)

# Replace "All Seasons" with ["Spring", "Summer", "Fall", "Winter"]
df_filtered["Season"] = df_filtered["Season"].apply(lambda x: ["Spring", "Summer", "Fall", "Winter"] if x == "All Seasons" else [x])

# Explode again to ensure each season gets its own row
df_final = df_filtered.explode("Season").reset_index(drop=True)
```

```python
df_final.to_excel("filtered_table_data_transformed.xlsx")
```
